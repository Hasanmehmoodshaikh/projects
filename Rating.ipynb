```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('fl.csv')
```

```python
df
```

```python
customer_rating = df.groupby('Product Name')['Customer Rating'].mean()
```

```python
font1 = {'family':'Arial','color':'darkgray','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
plt.figure(figsize = (10,6))
customer_rating.plot(kind = 'bar',color = 'black',linestyle = '--')
plt.title('Rating',fontdict = font1)
plt.xlabel('Product Name',fontdict = font2)
plt.ylabel('Customer Rating',fontdict = font2)
plt.grid(color ='gray',linestyle='--',linewidth = 0.5)

plt.show()
```

```python
customer_rating = df.groupby('Product Name')['Customer Rating'].mean()
print(customer_rating)
```

```python
plt.figure(figsize = (10,9))
customer_rating.plot(kind = 'pie',autopct = '%0.1f%%',startangle = 90,shadow = True)
plt.title('Rating Result')
plt.legend(title = 'Product Name')
plt.show()
```
