```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\hasan\\motoGP.csv")
```

```python
df.head(10)
```

```python
df = pd.DataFrame(df)
print(df)
```

```python
df.info()
```

```python
Jorge_Martin = df[df['rider_name'] == 'Jorge Martin']
y = Jorge_Martin['points']
x = Jorge_Martin['season']

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x,y, marker = 'o',ms = 9, mec = 'r', mfc = 'black')
plt.title('Jorge Martin Points',fontdict = font1,loc = 'left')
plt.xlabel('season',fontdict = font2)
plt.ylabel('points',fontdict = font2)
plt.annotate('Moto2', xy =(2019, 94), 
                xytext =(2019,30),  
                arrowprops = dict(facecolor ='red',
                                  edgecolor = 'black',linewidth = 1),)

plt.annotate('MotoGp', xy =(2021,111), 
                xytext =(2022,30),  
                arrowprops = dict(facecolor ='red', 
                                edgecolor = 'black',linewidth = 2, arrowstyle ='->',connectionstyle = 'arc3'),)

plt.annotate('Moto3', xy =(2016, 72), 
                xytext =(2018,260),  
                arrowprops = dict(facecolor ='red',
                                  edgecolor = 'black',linewidth = 1),)
plt.grid(color = 'blue')
plt.show()
```

```python
riders = df.groupby('rider_name')
```

```python
total_point = riders['points'].sum()
```

```python
rider_names = df['rider_name'].sort_values().unique()
```

```python
font1 = {'family':'Georgia','color':'darkgreen','size':20}
font2 = {'family':'Verdana','color':'red','size':15}
```

```python
f= plt.figure(figsize=(15,6))
plt.barh(rider_names,total_point,facecolor = 'black',edgecolor = 'red')
plt.title('Rider Points',fontdict = font1, loc ='left')
plt.xlabel('Score',fontdict = font2)
plt.ylabel('Riders Name',fontdict = font2)
plt.grid(axis = 'y',color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
total_win = riders['wins'].max()
total_podium = riders['podium'].max()
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
fig = plt.figure(figsize = (12,6))
plt.barh(rider_names,(total_win/total_podium*100),facecolor = 'gray',edgecolor = 'black')
plt.title('Podium And Winning',fontdict = font1, loc = 'left')
plt.xlabel('Percentage',fontdict = font2)
plt.ylabel('Rider Name',fontdict = font2)
plt.show()
```

```python
font1 = {'family':'Times New Roman','color':'#008080','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.figure(figsize = (10,6))
home.plot(kind ='bar',color = '#800000')
plt.title('Participate Country',fontdict = font1, loc = 'center')
plt.xlabel('home_country', fontdict = font2, color = '#008000')
plt.ylabel('races_participated', fontdict = font2, color = '#008000')
plt.grid(color ='black',linestyle='--',linewidth = 0.5)
plt.show()
```
