```python
 Flipkart = df.groupby('Product Name')['Price (₹)'].sum()
print(Flipkart)
```

```python
plt.figure(figsize = (10,6))
Flipkart.plot(kind = 'hist',facecolor = 'black', edgecolor = 'green' )
plt.title('Price Distribution')
plt.xlabel('Price (₹)')
plt.ylabel('Count')
plt.show()
```

```python
top_10_products = df.nlargest(10,'Price (₹)')
print(top_10_products)
```

```python
plt.figure(figsize = (10,6))
plt.plot(top_10_products['Price (₹)'],top_10_products['Product Name'])
plt.xlabel('Price')
plt.ylabel('Product Name')
plt.title('Top 10 Product by Price')
plt.grid(color = 'green')
plt.show()
```

```python
mean_rating = df['Rating (★)'].mean()
print(mean_rating)
```

```python
median_rating = df['Rating (★)'].median()
print(median_rating)
```

```python
std_rating = df['Rating (★)'].std()
print(std_rating)
```

```python
rating_freq = df['Rating (★)'].value_counts()
print(rating_freq)
```

```python
plt.figure(figsize = (10,7))
plt.hist(df['Rating (★)'],bins = 10, facecolor = 'yellow', edgecolor = 'black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Average Customer Rating (★)')
plt.show()
```

```python
unique_customer = df['Product Name'].nunique()
(unique_customer)
```

```python
unique_customer_list = df['Product Name'].unique()
(unique_customer_list)
```

```python
unique_customer_buyers = df.groupby('Number of Buyers')['Product Name'].nunique()
print(unique_customer_buyers)
```

```python
plt.figure(figsize = (15,5))
plt.plot(unique_customer_buyers.index,unique_customer_buyers.values, marker = 'o', mec = 'r', label = 'Number of Buyers')
plt.xlabel('Product Buyers')
plt.ylabel('Number of Buyers')
plt.title('Number of Buyers per Product')
plt.legend()
plt.show()
```

```python
total_sold = df.groupby('Product Name')['Total Sold'].sum()
print(total_sold)
```

```python
plt.figure(figsize = (10,6))
plt.plot(df.groupby('Seller')['Total Sold'].sum(), '+', color = 'r')
plt.xlabel('Seller')
plt.ylabel('Total Sold')
plt.title('Sales Trend')
plt.xticks(rotation = 90)
plt.grid(color = 'olive')
plt.show()
```

```python
product_comparison = df.groupby('Product Name')['Total Sold'].sum().reset_index() 
product_comparison.columns = ['Product Name','Total Sold']
product_comparison = product_comparison.sort_values(by = 'Total Sold',ascending = False)
print(product_comparison)
```

```python
sales_channal_analysis = df.groupby('Seller')['Total Sold'].sum()
(sales_channal_analysis)
```

```python
plt.figure(figsize = (10,6))
stock_by_category = df.groupby('Sub Category')['Available Stock'].sum().sort_values()
stock_by_category.plot(kind='barh', color='skyblue',edgecolor = 'black')
plt.title("Available Stock by Sub Category")
plt.grid(axis = 'y',color = 'r')
plt.show()
```

```python
plt.figure(figsize = (10,6))
stock_by_Discount = df.groupby('Main Category')['Discount (%)'].sum().sort_values()
stock_by_Discount.plot(kind='bar', color='darkblue',edgecolor = 'black')
plt.title("Discount (%) by Main Category")
plt.grid(axis = 'y',color = 'gray')
plt.show()
```

```python
return_policy = ['True','False']
count = [70,30]
```

```python
plt.figure(figsize = (8,9))
plt.pie(count,labels = return_policy,autopct = '%1.1f%%',startangle = 90,shadow = True)
plt.title('Return Policy')
plt.legend(title = 'True False ')
plt.show()
```
